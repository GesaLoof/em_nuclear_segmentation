```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# path to your CSV file
csv_path = "../training_log.csv" 

df = pd.read_csv(csv_path)
df["epoch"] = pd.to_numeric(df["epoch"], errors="coerce")
df = df.dropna(subset=["epoch"]).sort_values("epoch")

plt.figure(figsize=(8, 5))
plt.plot(df["epoch"], df["train_loss"], label="Training Loss", marker="o")
plt.plot(df["epoch"], df["val_loss"], label="Validation Loss", marker="o")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title(f"Training and Validation Loss (epochs {int(df['epoch'].min())}â€“{int(df['epoch'].max())})")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
